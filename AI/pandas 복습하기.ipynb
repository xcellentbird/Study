pandas의 read_csv를 사용하여 csv(' ","로 나눠진 데이터셋 파일 ')를 읽어들일 수 있다. 반환되는 데이터 타입은 pd.DataFrame 타입이다.

첫번째 파라미터로 csv파일의 위치(str)를 받으며, sep(데이터를 나누는 문자)를 설정함으로서 csv 외의 파일도 불러올 수 있을 것으로 보인다. encoding 파라미터를 수정하여 들어오는 데이터의 인코딩 형식으로 데이터를 읽어올 수 있다.

```python
import pandas as pd

data = pd.read_csv('../input/korean-real-estate-transaction-data/Apart Deal.csv', encoding='utf-8', na_values=' ')
```

```python
data.iloc[2]
```

- loc은 DataFrame의 column이름이나 index이름으로 데이터를 불러온다.

```python
data.loc[2:5, '거래일': '지번']
```

- 조건형으로 원하는 데이터만을 불러올 수도 있다.

- describe와 count, mean, std, median 등을 통해서 수치형 데이터의 특징을 구할 수도 있다.

```python
data.describe()
```

```python
data['층'].median()
```

- Series의 unique()를 이용하면 해당 칼럼의 고유값들을 알 수 있다. 그리고 value_counts()를 사용하면, 해당 데이터가 몇 개 있는지 확인 가능하다.

```python
data['아파트'].nunique(), data['아파트'].unique()
```

```python
data['아파트'].value_counts()
```

- 아래와 같이 select_dtypes를 이용하면 원하는 데이터형이 보관된 칼럼만 확인할 수 있다.

```python
data.select_dtypes('object')
```

- Series에 isnull()를 이용하면, 해당 칼럼의 데이터가 nan값인지(bool형태) 알 수 있다. 더해서 sum()을 이용하면 해당 칼럼의 nan 데이터 개수를 알 수 있다.

```python
data['층'].isnull().sum()
```

- map과 apply를 이용해서 데이터를 mapping 할 수 있다. 
- map은 해당 칼럼 Series의 데이터를 함수(or lambda)를 이용하여 바꿀 수 있다.
- apply는 DataFrame이나 Series의 데이터를 확인하면서 mapping 할 수 있다. axis 설정을 통해 칼럼 방향, 인덱스 방향 설정이 가능하다.

```python
data.loc[:100,'건축년도'].map(lambda x: x-int(data['건축년도'].mean()))
```

```python
def remean_points(row):
    row['건축년도'] = row['건축년도'] - int(data['건축년도'].mean())
    return row
```

```python
data.iloc[:100].apply(remean_points, axis='columns')
```

- groupby를 이용하면 해당 칼럼에 대한 다른 칼럼 특징(mean, median, std, 등)을 불러올 수 있다.

```python
data[:1000].groupby('아파트').describe()
```

```python
data[:1000].groupby('아파트')['거래금액'].agg([len, min, max])
```

- sort_values()를 이용하면, 원하는 형태로 데이터 정렬이 가능하다.
- 또한, sort_index()를 통해 index 순서로 정렬도 가능하다.

```python
data[:1000].groupby('아파트')['거래금액'].mean().sort_values()
```

```python
data[:1000].sort_values(by=['거래금액', '층'], ascending=False)
```

- dropna()를 이용하면 nan값을 뺀 데이터셋을 가져올 수 있다

```python
data.dropna()
```

- pd.to_datetime을 이용하면, dtype을 datetime으로 바꿀 수 있다. 
- datetime형태로 바꾸게 되면, 날짜나 시간 데이터를 쉽게 다룰 수 있다.

```python
data['거래일'] = pd.to_datetime(data['거래일'])
data.info()
```
